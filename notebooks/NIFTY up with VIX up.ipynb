```python
import numpy as np
import pandas as pd

import datetime
from niftyutils import load_nifty_data
```

```python
DATA_DIR = './../data/'
```

```python
start_date = datetime.datetime(2010,9,30)
end_date = datetime.datetime(2020,9,30)
nifty_data = load_nifty_data(start_date,end_date)
```

```python
nifty_data['pct_change'] = (nifty_data['Close']/nifty_data['Close'].shift(1) - 1)*100
nifty_data = nifty_data.dropna()
nifty_data
```

```python
vix_data = pd.read_csv(DATA_DIR+'vix_data.csv',index_col=['Date '],parse_dates=True)
```

```python
nifty_vix_data = pd.concat([nifty_data['pct_change'].rename('nifty change'),vix_data['% Change'].rename('vix change')], axis=1, join = 'inner')
```

```python
nifty_vix_data[(nifty_vix_data['nifty change'] > 0.0) & (nifty_vix_data['vix change'] > 5)]
```
